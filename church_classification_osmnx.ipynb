# Geospatial Data Classification: Churches, Belonging, and Public OSM Tags

This notebook adapts the original geospatial data classification lab to examine churches in Seattle through a critical GIS lens.

The guiding idea comes from Martin Luther King Jr.'s observation that Sunday morning worship has often reflected racial segregation in the United States. Here, the goal is not to determine the actual racial identity of congregants. Instead, this notebook asks a more limited and methodologically careful question:

> Which churches are publicly legible in OpenStreetMap as Asian, Black, African, or African-diasporic religious spaces, and which churches remain outside those visible signals?

The notebook creates two side-by-side maps using the same legend:

- **Left map:** Asian Church classification
- **Right map:** Black Church classification

Both maps use two categories:

| Category | Meaning |
|---|---|
| `Spiritual and Belonging` | A church whose public OSM name or tags signal the selected community identity |
| `Unfamiliar, Yet Welcoming` | Other Christian churches that may still be welcoming, but do not visibly signal that identity in OSM |

This is a classification of **public data signals**, not a demographic claim about church members.

```python
# Import libraries
import re

import osmnx as ox
import geopandas as gpd
import pandas as pd
import folium
from folium.plugins import DualMap
```

## Step 1: Retrieve Church Data from OpenStreetMap

OSMnx retrieves features based on OpenStreetMap tags. For churches, the most useful starting point is:

- `amenity = place_of_worship`
- `building = church`

After downloading, we filter toward Christian churches using available OSM fields such as `religion`, `building`, and `name`.

Important limitation: OSM usually does **not** have a formal ethnicity tag for churches. Therefore, the later classification relies on public textual signals in fields such as `name`, `denomination`, `operator`, and `description`.


```python
# Download place-of-worship and church features from OSM for Seattle
place_name = "Seattle, Washington, USA"

# These tags are intentionally broad. OSMnx returns features matching any of them.
tags = {
    "amenity": "place_of_worship",
    "building": "church"
}

pois = ox.features_from_place(place_name, tags)

# Keep a copy and make geometry usable for point-based Folium markers.
# Some churches are mapped as building polygons, so representative_point() avoids dropping them.
pois = pois.reset_index()
pois = pois[pois.geometry.notna()].copy()
pois["marker_geometry"] = pois.geometry.representative_point()

# OSM columns vary by place. Ensure these fields exist before filtering/classifying.
for col in ["amenity", "religion", "building", "name", "denomination", "operator", "description"]:
    if col not in pois.columns:
        pois[col] = pd.NA

# Filter to likely Christian churches.
name_text = pois["name"].fillna("").astype(str)
religion_text = pois["religion"].fillna("").astype(str).str.lower()
building_text = pois["building"].fillna("").astype(str).str.lower()

churches = pois[
    (religion_text == "christian") |
    (building_text == "church") |
    (name_text.str.contains(r"church|chapel|ministry|fellowship|cathedral|parish", case=False, na=False))
].copy()

print(f"Downloaded {len(pois)} worship-related OSM features.")
print(f"Filtered to {len(churches)} likely Christian church features.")
churches[["name", "amenity", "religion", "denomination", "building"]].head(10)
```

```python
# Inspect available OSM columns and a sample of church names
print("Available columns:")
print(sorted(churches.columns.tolist()))

churches[["name", "religion", "denomination", "operator", "description"]].dropna(how="all").head(20)
```

## Step 2: Define Community-Signal Tags

Because OSM does not reliably record church ethnicity, this notebook uses **visible public signals**. The search terms below are applied to OSM text fields.

For the Asian church map, terms include Asian national, ethnic, linguistic, and regional identifiers such as `Korean`, `Chinese`, `Vietnamese`, `Filipino`, `Mandarin`, and `Cantonese`.

For the Black church map, terms include historically Black denominations and Black church tradition and African community identifiers such as `AME`, `COGIC`, `National Baptist`, `Ethiopian`, `Eritrean`, `Nigerian`, `Ghanaian`, `Haitian`, and `Jamaican`.

This approach is intentionally limited. It detects what is publicly visible in OSM, not the full social identity of a congregation.


```python
# Classification terms for visible public identity signals in OSM text fields
TEXT_FIELDS = ["name", "denomination", "operator", "description"]

ASIAN_PATTERNS = [
    r"\basian\b",
    r"\bkorean\b", r"\bkorea\b",
    r"\bchinese\b", r"\bchina\b",
    r"\bvietnamese\b", r"\bvietnam\b",
    r"\bjapanese\b", r"\bjapan\b",
    r"\bfilipino\b", r"\bphilippine\b", r"\bphilippines\b",
    r"\bhmong\b",
    r"\bcambodian\b", r"\bcambodia\b", r"\bkhmer\b",
    r"\blao\b", r"\blaotian\b",
    r"\bthai\b", r"\bthailand\b",
    r"\btaiwanese\b", r"\btaiwan\b",
    r"\bindian\b", r"\bindia\b",
    r"\bindonesian\b", r"\bindonesia\b",
    r"\bmandarin\b", r"\bcantonese\b",
]

BLACK_PATTERNS = [
    # Explicit racial / community identity
    r"\bblack\b",
    r"\bafrican american\b",
    r"\bafro[- ]american\b",

    # Historically Black denominations / traditions
    r"\bafrican methodist episcopal\b",
    r"\ba\.?m\.?e\.?\b",
    r"\bame zion\b",
    r"\bafrican methodist episcopal zion\b",
    r"\bchurch of god in christ\b",
    r"\bcogic\b",
    r"\bnational baptist\b",
    r"\bprogressive national baptist\b",
    r"\bmissionary baptist\b",

    # African and Black community identity
    r"\bafrican\b",
    r"\bethiopian\b", r"\bethiopia\b", r"\bethio\w*\b", r"\babyssinian\b",
    r"\beritrean\b", r"\beritrea\b",
    r"\bghanaian\b", r"\bghana\b",
    r"\bnigerian\b", r"\bnigeria\b",
    r"\bkenyan\b", r"\bkenya\b",
    r"\bsomali\b", r"\bsomalia\b",
    r"\bsudanese\b", r"\bsudan\b",
    r"\bcongolese\b", r"\bcongo\b",
    r"\bcameroonian\b", r"\bcameroon\b",
    r"\bliberian\b", r"\bliberia\b",
    r"\bsierra leonean\b", r"\bsierra leone\b",

    # Caribbean / Afro-diasporic identity
    r"\bcaribbean\b",
    r"\bjamaican\b", r"\bjamaica\b",
    r"\bhaitian\b", r"\bhaiti\b",
    r"\btrinidad\b", r"\btobago\b",
    r"\bbarbadian\b", r"\bbarbados\b",
]


def combined_text(row):
    return " ".join(str(row.get(col, "")) for col in TEXT_FIELDS if pd.notna(row.get(col, pd.NA))).lower()


def first_matching_pattern(text, patterns):
    for pattern in patterns:
        if re.search(pattern, text, flags=re.IGNORECASE):
            return pattern
    return None


def classify_by_patterns(row, patterns):
    text = combined_text(row)
    match = first_matching_pattern(text, patterns)
    if match:
        return "Spiritual and Belonging", match
    return "Unfamiliar, Yet Welcoming", None


churches[["asian_map_category", "asian_match"]] = churches.apply(
    lambda row: pd.Series(classify_by_patterns(row, ASIAN_PATTERNS)), axis=1
)
churches[["black_map_category", "black_match"]] = churches.apply(
    lambda row: pd.Series(classify_by_patterns(row, BLACK_PATTERNS)), axis=1
)

print("Asian map category counts:")
print(churches["asian_map_category"].value_counts())
print("\nBlack map category counts:")
print(churches["black_map_category"].value_counts())
```

## Step 3: Review Matched Churches Before Mapping

Before mapping, inspect the matched records. This is an important critical GIS step because automated classification can misread names, miss important communities, or overgeneralize from partial tags.

If a match looks wrong, revise the pattern list above. If important churches are missing, that may reflect a limitation of OSM tagging rather than the absence of those communities.


```python
# Review matched churches for both map logics
asian_matches = churches[churches["asian_map_category"] == "Spiritual and Belonging"][
    ["name", "denomination", "asian_match"]
].sort_values("name")

black_matches = churches[churches["black_map_category"] == "Spiritual and Belonging"][
    ["name", "denomination", "black_match"]
].sort_values("name")

print("Asian Church visible-signal matches:")
display(asian_matches)

print("Black Church visible-signal matches:")
display(black_matches)
```

## Step 4: Create the Side-by-Side Counter-Map

The two maps use the same legend and categories. What changes is the subject of classification:

- **Left:** Asian Church map
- **Right:** Black Church map

The category `Unfamiliar, Yet Welcoming` is not meant as a negative label. It means that the church does not visibly signal the selected community identity in the available OSM tags, while still being interpreted as potentially open or welcoming.


```python
# Create side-by-side map comparison
color_map = {
    "Spiritual and Belonging": "green",
    "Unfamiliar, Yet Welcoming": "blue"
}

legend_html = """
<div style="position: fixed; bottom: 32px; left: 50px; z-index: 9999; background: white;
            padding: 12px 14px; border: 1px solid #999; border-radius: 6px;
            font-size: 13px; box-shadow: 0 2px 8px rgba(0,0,0,0.2);">
  <strong>Legend</strong><br>
  <span style="color: green; font-size: 18px;">●</span> Spiritual and Belonging<br>
  <span style="color: blue; font-size: 18px;">●</span> Unfamiliar, Yet Welcoming
</div>
"""

center = [47.61, -122.33]
dual_map = DualMap(location=center, zoom_start=12, tiles="CartoDB positron")

# Add titles to each map pane.
title_left = """
<div style="position: fixed; top: 72px; left: 70px; z-index: 9999; background: white;
            padding: 10px 12px; border: 1px solid #aaa; border-radius: 6px;
            font-weight: bold; box-shadow: 0 2px 8px rgba(0,0,0,0.18);">
  Asian Church Map
</div>
"""

title_right = """
<div style="position: fixed; top: 72px; right: 70px; z-index: 9999; background: white;
            padding: 10px 12px; border: 1px solid #aaa; border-radius: 6px;
            font-weight: bold; box-shadow: 0 2px 8px rgba(0,0,0,0.18);">
  Black Church Map
</div>
"""

# Left map: Asian Church classification
for _, row in churches.iterrows():
    category = row["asian_map_category"]
    color = color_map.get(category, "gray")
    geom = row["marker_geometry"]
    popup = folium.Popup(
        f"""
        <b>{row.get('name', 'Unnamed church')}</b><br>
        <b>Map:</b> Asian Church<br>
        <b>Category:</b> {category}<br>
        <b>Matched signal:</b> {row.get('asian_match') or 'No visible Asian signal in selected OSM fields'}<br>
        <b>Denomination:</b> {row.get('denomination') or 'Unknown'}
        """,
        max_width=320
    )
    folium.CircleMarker(
        location=(geom.y, geom.x),
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.8,
        popup=popup
    ).add_to(dual_map.m1)

# Right map: Black Church classification
for _, row in churches.iterrows():
    category = row["black_map_category"]
    color = color_map.get(category, "gray")
    geom = row["marker_geometry"]
    popup = folium.Popup(
        f"""
        <b>{row.get('name', 'Unnamed church')}</b><br>
        <b>Map:</b> Black Church<br>
        <b>Category:</b> {category}<br>
        <b>Matched signal:</b> {row.get('black_match') or 'No visible Black church signal in selected OSM fields'}<br>
        <b>Denomination:</b> {row.get('denomination') or 'Unknown'}
        """,
        max_width=320
    )
    folium.CircleMarker(
        location=(geom.y, geom.x),
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.8,
        popup=popup
    ).add_to(dual_map.m2)

# Add shared map elements.
dual_map.get_root().html.add_child(folium.Element(legend_html))
dual_map.get_root().html.add_child(folium.Element(title_left))
dual_map.get_root().html.add_child(folium.Element(title_right))

dual_map
```

## Final Reflection

This counter-map uses OSM data to explore how religious belonging becomes visible or invisible in public geospatial data. The two classifications do not claim to identify the race or ethnicity of church members. Instead, they classify churches according to visible textual signals in volunteered geographic information.

This matters because OSM can make some communities easier to see than others. A church with a name such as "Korean," "Ethiopian," "AME," or "COGIC" may be classified as `Spiritual and Belonging` in one of the two maps. Another church may serve immigrant, Asian, Black, or multiethnic communities but remain classified as `Unfamiliar, Yet Welcoming` because those identities are not visible in OSM tags.

The classification therefore reveals both presence and absence: where community identity is publicly named, and where it is hidden, untagged, ambiguous, or not reducible to map data.
